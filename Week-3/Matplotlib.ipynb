```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('dark_background')
```

## Bar Plot

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.bar(subjects, marks)
plt.show()
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(subjects, marks, color = colors)
plt.show()
plt.barh(subjects, marks, color = colors)
plt.show()
```

```python
df = pd.read_csv(r'E:\Software Sub Team Weekly Tasks\Week-3\DataSets\Sample_3.csv')
df.head()
```

```python
city_rating = df.groupby('City')['Rating'].mean()

plt.bar(city_rating.index, city_rating.values)
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.title("Average Customer Rating by City")
plt.show()

sales_df = df.groupby('Product line')['Total'].sum()

plt.bar(sales_df.index, sales_df.values)
plt.xticks(rotation=45)
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.title("Total Sales per Product Line")
plt.tight_layout()
plt.show()

payment_df = df['Payment'].value_counts()

plt.bar(payment_df.index, payment_df.values)
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.title("Payment Method Distribution")
plt.show()
```

## Pie Plot

```python
plt.figure(figsize=(7,7))

df['Customer type'].value_counts().plot(
    kind='pie',
    colors=['blue', 'green'],
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Customer Type Distribution (Pie Chart)")
plt.ylabel("")    # remove default text
plt.show()


payment_counts = df['Payment'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index,
        colors=['red', 'blue', 'green'], autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()


gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Unit price'], df['Quantity'])
plt.xlabel("Unit Price")
plt.ylabel("Quantity")
plt.title("Scatter Plot: Unit Price vs Quantity")
plt.show()
```
