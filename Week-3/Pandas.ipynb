## Handling Null Or Missing Values

```python
df2 = pd.read_csv(r'E:\Software Sub Team Weekly Tasks\Week-3\DataSets\Sample_2.csv')
#print(df2)
print(df2.head())
```

```python
df2.isnull()
```

```python
print(df2.isnull().sum())
print("Total Null Values:",df2.isnull().sum().sum())
```

```python
## Droping rows with Null Values
print("Shape with Null value:",df2.shape)
# removing null and storing in df3 
df3 = df2.dropna() # df2.dropna(asix = 0)- > drop rows with null, for 1 it will drop a column with null value
print("df3 shape:", df3.shape)

# df2.dropma(how ='any') # if any row value is null then remove that row
# df2.dropma(how ='all') # if all row value is null then remove that row
# df2.dropna(inplace = True) # it will remove the Null rows and replace the orginal dataframe 
```

```python
## filling the null values
print(df2.fillna(0)) # fill the null values with 0
# print(df2.fillna({'Physics':'none', 'Chemistry':0, 'Maths':30}) )
#print(df2.ffill()) # fill with previous value
#print(df2.ffill(axis= 1)) # fill with previous column value
#print(df2.ffill()) #fill with next value
print(df2['Physics'].fillna(value = df2['Physics'].mean())) # fill with mean value

```
